```python
import pandas as pd
import numpy as np
```

```python
seller_data=pd.read_csv('../FeatureEng/data_under_observation_count.csv')
seller_data = seller_data.drop(columns = ['Unnamed: 0'])
seller_data.head()
```

```python
brazil_gdp=pd.read_csv('../FeatureEng/brazilEcon.csv')
brazil_gdp.rename(columns={'year':'transaction_year','month':'transaction_month'}, inplace=True)
brazil_gdp.head()
```

```python
#Joining above 2 tables

seller_brazil_gdp=pd.merge(seller_data, brazil_gdp, how='inner', left_on=['transaction_year','transaction_month'],right_on=['transaction_year','transaction_month']) #full outer join based on seller id
print(seller_brazil_gdp)
```
